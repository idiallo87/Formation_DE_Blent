# Les boucles

```python
liste = [1, 5, 9, 12]
for x in liste: # x est l'élément itéré sur la liste
    print(x)
```

```python
liste = [1, 5, 9, 12]
somme = 0

for x in liste: # x est l'élément itéré sur la liste
    somme += x # Cette instruction signifie que l'on ajoute x à la variable somme
    
print("La somme vaut", somme)
```

# La fonction range

```python
for i in range(50):
    print(i)
```

```python
matrice = [["Pierre", 34, "Paris"],
           ["Jean", 69, "Bordeaux"],
           ["Aurélie", 28, "Tours"]]

for i in range(3): # Pour chaque personne
    for j in range(3): # Pour chaque information
        print("Individu", i + 1, "information n°", j + 1, ":", matrice[i][j])
```

# La boucle while

```python
import random

nombre = random.randint(0, 100)

while nombre <= 80:
    print(nombre)
    nombre = random.randint(0, 100)
    
print("Fin")
```

```python
nombre = random.randint(0, 100)

while nombre <= 95:
    print(nombre)
    nombre = random.randint(0, 100)
    
print("Fin")
```

```python
nombre = random.randint(0, 100)
tirages = 0

while nombre <= 95 and tirages < 5:
    print(nombre)
    tirages += 1 # Ajoute 1 à la variable tirages
    nombre = random.randint(0, 100)

print("Fin")
```

```python
nombre = random.randint(0, 100)
tirages = 0

while nombre <= 95 and tirages < 5:
    print(nombre)
    tirages += 1 # Ajoute 1 à la variable tirages
    nombre = random.randint(0, 100)
    
if tirages == 5:
    print("Fin stoppée")
else:
    print("Fin normale")
```

# Les dictionnaires

```python
dico = dict({
    'prenom': "Jean",
    'nom': "Dupont",
    'age': 35
})

print(dico['prenom'], dico['nom'], "a", dico['age'], "ans.")
```

```python
dico['ville'] = "Paris"
```

```python
mondico = dict({})

cles = ['prenom', 'ville', 'age']
valeurs = ["Jean", "Dupont", 35]

for i in range(3):
    mondico[cles[i]] = valeurs[i]
        
print(mondico)
```

```python
for cle in dico:
    print("Clé :", cle)
```

```python
for cle in dico.values():
    print("Valeurs :", cle)
```
